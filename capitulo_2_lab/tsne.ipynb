```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.manifold import TSNE
```

```python
iris = datasets.load_iris()
X=iris.data
y=iris.target
```

```python
tsne = TSNE(n_components=2, n_iter=1000, random_state=123)
iris_t= tsne.fit_transform(X)
```

```python
plt.scatter(iris_t[y==0, 0], iris_t[y==0, 1], marker='o')
plt.scatter(iris_t[y==1, 0], iris_t[y==1, 1], marker='*')
plt.scatter(iris_t[y==2, 0], iris_t[y==2, 1], marker='s')

plt.xlabel('Primera variable transformada')
plt.ylabel('Segunda variable transformada')
plt.title('Representación espacio transformado')
plt.savefig("datos_transformados por el t-SNE.pdf")
plt.show()
```
