```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
np.random.seed(111)
X,y=make_blobs(n_samples=150,cluster_std=0.50, centers=4)
print(X.shape,y.shape)
```

```python
plt.figure(0)
plt.grid(True)
plt.scatter(X[:,0],X[:,1])
plt.savefig("k_means_patrones.pdf")
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
grupos=KMeans(n_clusters=2)
grupos.fit(X)
```

```python
centros=grupos.cluster_centers_
labels=grupos.labels_
plt.scatter(X[labels==0, 0], X[labels==0, 1], marker='o')
plt.scatter(X[labels==1, 0], X[labels==1, 1], marker='s')
plt.scatter(centros[:,0],centros[:,1],c='black', marker='*')
plt.savefig("k_means_2_grupos.pdf")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
centros= range(2,10)
ind=[]

for numero in centros:
    modelo = KMeans(n_clusters=numero)
    labels = modelo.fit_predict(X)
    silhouette_medio = silhouette_score(X,labels)
    ind.append(silhouette_medio)
    
fig, ax = plt.subplots(1, 1, figsize=(6, 4))
ax.plot(centros, ind, marker='o')
ax.set_title("Índices Silhouette")
ax.set_xlabel('Número grupos')
ax.set_ylabel('Media índices silhouette');
plt.savefig("indice_silhoutte.pdf")
```
