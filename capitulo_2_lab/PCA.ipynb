```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import mpl_toolkits.mplot3d
```

```python
iris = datasets.load_iris()
X=iris.data
y=iris.target
x_e= StandardScaler().fit_transform(X)

pca = PCA()
pca.fit(x_e)

cumsum = 100*np.cumsum(pca.explained_variance_ratio_)
print("varianza acumulada=" ,cumsum)

dim = np.arange(len(cumsum))+1
plt.bar(dim,cumsum);
plt.xlabel('Dimensiones')
plt.ylabel('% Varianza Explicada')

plt.xticks([1,2,3,4])
plt.ylim([0,110])
plt.savefig("porcentaje_varianza.pdf")
```

```python
#plt.scatter(X[:,0], X[:,1], s=35, c=y, cmap=plt.cm.brg)



plt.scatter(X[y==0, 0], X[y==0, 1], marker='o')
plt.scatter(X[y==1, 0], X[y==1, 1], marker='*')
plt.scatter(X[y==2, 0], X[y==2, 1], marker='s')

plt.xlabel('Longitud sépalo')
plt.ylabel('Anchura sépalo')
plt.title('Longitud sépalo vs. Anchura sépalo')
plt.savefig("datos_originales_iris.pdf")
plt.show()
```

```python
pca = PCA(n_components=2)
p_c = pca.fit_transform(x_e)

plt.scatter(p_c[y==0, 0], p_c[y==0, 1], marker='o')
plt.scatter(p_c[y==1, 0], p_c[y==1, 1], marker='*')
plt.scatter(p_c[y==2, 0], p_c[y==2, 1], marker='s')

plt.xlabel('Primera componente')
plt.ylabel('Segunda componente')
plt.title('Representación espacio transformado')
plt.savefig("datos_PCA_iris.pdf")
plt.show()
```
