```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('geyser')
df.head()
```

```python
fig, ax = plt.subplots()

plt.subplot(2,2,1)
plt.hist(df.duration, bins=5)
plt.title('Histograma-5 bins') 
plt.xlabel('Duración') 

plt.subplot(2,2,2)
plt.hist(df.duration, bins=10)
plt.title('Histograma-10 bins') 
plt.xlabel('Duración') 

plt.subplot(2,2,3)
plt.hist(df.duration, bins=20)
plt.title('Histograma-20 bins') 
plt.xlabel('Duración') 

plt.subplot(2,2,4)
plt.hist(df.duration, bins=100)
plt.title('Histograma-100 bins') 
plt.xlabel('Duración') 

fig.tight_layout()

plt.savefig("Histogramas.pdf")
plt.show()
```
